```python
census_econ = pd.read_csv(os.path.join(data_dir, "./employment_data/detailed_employment_data.csv"))

# Define the list of unwanted states
unwanted_states = [
    'Dakota', 'District Of Columbia', 'Indian', 'New', 'Persons in the Military'
]

weird_states = [
    'Dakota', 'Indian', 'New', 'Persons in the Military'
]

census_econ_weird = census_econ[census_econ['STATE'].isin(weird_states)]

print(census_econ_weird.shape)
```

```python
# Filter out the unwanted states
census_econ_clean = census_econ[~census_econ['STATE'].isin(unwanted_states)]
```

```python
census_econ_1880 = census_econ_clean[census_econ_clean['YEAR'] == 1880]

# List of states that existed by 1880
states_1880 = pd.DataFrame({'STATE': [
    'Alabama', 'Arkansas', 'California', 'Colorado', 'Connecticut', 'Delaware',
    'Florida', 'Georgia', 'Illinois', 'Indiana', 'Iowa', 'Kansas', 'Kentucky',
    'Louisiana', 'Maine', 'Maryland', 'Massachusetts', 'Michigan', 'Minnesota',
    'Mississippi', 'Missouri', 'Nebraska', 'Nevada', 'New Hampshire',
    'New Jersey', 'New York', 'North Carolina', 'Ohio', 'Oregon',
    'Pennsylvania', 'Rhode Island', 'South Carolina', 'Tennessee', 'Texas',
    'Vermont', 'Virginia', 'West Virginia', 'Wisconsin'
]})

# Filter census data for the year 1880
census_econ_1880 = census_econ[census_econ['YEAR'] == 1880]

# Perform a full outer merge
merged_states = states_1880.merge(census_econ_1880[['STATE']], on='STATE', how='outer', indicator=True)

# Check states absent in census_econ_1880
absent_states = merged_states[merged_states['_merge'] == 'left_only']['STATE']
print("States absent in census_econ_1880:")
print(absent_states)

# Check states not supposed to be there but present in census_econ_1880
extra_states_1880 = merged_states[merged_states['_merge'] == 'right_only']['STATE']
print("\nStates not supposed to be there but present in census_econ_1880:")
print(extra_states_1880)
```
